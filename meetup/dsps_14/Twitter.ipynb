```python
import json

# Enter your keys/secrets as strings in the following fields
credentials = {}
credentials['CONSUMER_KEY'] = "<myConsumerKey>"
credentials['CONSUMER_SECRET'] = "<myConsumerSecret>"
credentials['ACCESS_TOKEN'] = "<myAccessToken>"
credentials['ACCESS_SECRET'] = "<myAccessSecret"

# Save the credentials to file
with open("twitter_credentials.json", "w") as file:
    json.dump(credentials, file)
```

```python
#Fetching the Tweets

from twython import Twython
import json

#with open("twitter_credentials.json", "r") as file:
  #creds = json.load(file)

python_tweets = Twython(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])

#Query
query = {'q': 'gamestop',
         'result_type': 'popular',
         'count': 10000,
         'lang': 'en',
        }
```

```python
import pandas as pd

#search
dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': []}
for status in python_tweets.search(**query)['statuses']:
  dict_['user'].append(status['user']['screen_name'])
  dict_['date'].append(status['created_at'])
  dict_['text'].append(status['text'])
  dict_['favorite_count'].append(status['favorite_count'])

#pandas for easier handling
df = pd.DataFrame(dict_)
df.sort_values(by='favorite_count', inplace=True, ascending=False)
df.head(10000)
```

```python
#Output formatting
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
display(df)
```

```python
import tweepy
import csv

consumer_key    = creds['CONSUMER_KEY']
consumer_secret = creds['CONSUMER_SECRET']
access_key      = creds['ACCESS_TOKEN']
access_secret   = creds['ACCESS_SECRET']

def get_all_tweets(screen_name):
  #authorize access
  auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
  auth.set_access_token(access_key, access_secret)
  api = tweepy.API(auth)

  #for all our tweets fetched
  alltweets = []
  
  new_tweets = api.user_timeline(screen_name = screen_name, tweet_mode="extended", count=200)

  alltweets.extend(new_tweets)
  
  oldest = alltweets[-1].id - 1

  #process each tweet
  while len(new_tweets) > 0:
    print(f"getting tweets before {oldest}")
    new_tweets = api.user_timeline(screen_name = screen_name, tweet_mode="extended", count=200, max_id=oldest)
    alltweets.extend(new_tweets)
    oldest = alltweets[-1].id -1
    print(f'...{len(alltweets)} tweets downloaded so far')

  #writing to csv
  outtweets = [[tweet.id_str, tweet.created_at, tweet.full_text] for tweet in alltweets]
  with open(f'new_{screen_name}_tweets.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(["id", "created_at", "text"])
    display(outtweets)
    writer.writerows(outtweets)
  pass


if __name__ == '__main__':
  get_all_tweets("gamestop")
```
